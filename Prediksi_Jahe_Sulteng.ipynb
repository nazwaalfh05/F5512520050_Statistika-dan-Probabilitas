```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

nama_file = r"D:\ORION\Matkul\Data_Jahe_Sulteng.xlsx" 
df = pd.read_excel(nama_file, header=1)
df.columns = df.columns.str.strip()

print("Kolom yang terdeteksi sekarang:", df.columns.tolist())
X = df['Luas panen jahe']
Y = df['Produksi Jahe']

print("\nData Jahe Sulteng berhasil dimuat ke Python!")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', alpha=0.7, label='Data Riil BPS (2018-2025)')

X_garis = np.linspace(X.min(), X.max(), 100)
Y_garis = 0.3349 * X_garis + 22814 
plt.plot(X_garis, Y_garis, color='red', linestyle='--', linewidth=2, 
         label='Model Regresi ML (y = 0,3349x + 22814)')

plt.title('Grafik Tren Prediksi Produksi Jahe Provinsi Sulawesi Tengah (2018-2025)', fontsize=14, pad=15)
plt.xlabel('Luas Panen Jahe (Hektar)', fontsize=12)
plt.ylabel('Total Produksi Jahe (Kilogram)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)

plt.show()
```

```python
def prediksi_panen_jahe(luas_lahan_baru):
    hasil = (0.3349 * luas_lahan_baru) + 22814
    return round(hasil, 2)
input_lahan_baru = 150000 
hasil_tebakan = prediksi_panen_jahe(input_lahan_baru)

print("==============================================")
print(f"Jika Target Luas Lahan Panen : {input_lahan_baru:,} Hektar")
print(f"Maka Estimasi Hasil Produksi : {hasil_tebakan:,} Kuintal")
print("==============================================")
```
